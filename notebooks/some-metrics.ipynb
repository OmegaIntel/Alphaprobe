```python
# extract values for some metrics

from api.doc_parser.doc_utils import flatten_dict_once, dict_from_summary_json, extract_key_val_from_dict

from glob import glob
import pandas as pd
```

```python
metrics = {
    'report_title': 'report_title',
    'revenue_cagr_value_projected': 'revenue_cagr_projected.revenue_cagr_value',
    'revenue_dollars': 'revenue_dollars',
    'profit_margins_percentage': 'profit_margins.profit_margins_percentage',
    'segment_percentage': 'market_segmentation.segment_percentage',
}
```

```python
filenames = glob('rag_outputs/*/section_summaries.json')
len(filenames)
```

```python
for_df = []

for filename in filenames:
    with open(filename) as f:
        text = f.read()

    dd = dict_from_summary_json(text)
    dd = flatten_dict_once(dd)
    dd = flatten_dict_once(dd)

    out = {}
    for metric_name, metric_key in metrics.items():
        val = extract_key_val_from_dict(dd, metric_key)
        out[metric_name] = val
    for_df.append(out)

df = pd.DataFrame(for_df)
df.head(10)
```

```python
df.to_csv('revenue-and-percentage-metrics.csv', index=False)
```
